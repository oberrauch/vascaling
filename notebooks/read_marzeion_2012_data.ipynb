```python
import os

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# define column names
c = [f'sum_{i}' for i in range(1,20)]
c.extend([f'std_{i}' for i in range(1,20)])
```

```python
fig = plt.figure()

# locate directory and list all files
dir_path = '/Users/oberrauch/Downloads/tc-6-1295-2012-supplement/'
for f in os.listdir(dir_path):

    # subset for files containing regional accumulated SLR contributionm
    if 'regional_accum' in f:
        f_path = os.path.join(dir_path, f)
        ds_ = pd.read_csv(f_path, delim_whitespace=True, index_col=0, header=0)
        ds_.columns = c
        # convert into volume
        plt.plot(ds_['sum_11'].loc[2000:2100]/ds_['sum_11'].loc[2000])
```
