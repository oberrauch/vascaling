```python
import os
import numpy as np
import pandas as pd
import xarray as xr
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# plot volume
plt.figure(figsize=[10,10])
dir_path = '../data/'
f_name = 'no-bias-sensitivity.nc'
with xr.open_dataset(os.path.join(dir_path, f_name)).isel(rgi_id=0, normalized=0) as ds:
    ds.volume.plot.line(x='time')
f_name = 'bias-sensitivity.nc'
with xr.open_dataset(os.path.join(dir_path, f_name)).isel(rgi_id=0, normalized=0) as ds:
    ds.volume.plot.line(x='time')
```
