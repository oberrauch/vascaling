```python
import numpy as np
from numpy import pi, sin
import matplotlib.pyplot as plt
%matplotlib inline
```

```python


x = np.linspace(0, 4*pi, 360)
with plt.xkcd():
    
    
    
    
    plt.plot(x, 0.5*sin(x))
    plt.axhline(0, c='k', ls=':')
    plt.ylim([-1.5,1.5])
    ax = plt.gca()
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.set_xticks([])
    ax.set_xlabel('Time')
    ax.set_yticks([])
    ax.set_ylabel('Amplitude')
    plt.text(0, -1.2, r'f(t) = $\frac{1}{2}$sin(t)', c='C0')
    plt.savefig('/Users/oberrauch/Desktop/slides/pds1.pdf')

    plt.figure()
    plt.plot(x, 0.5*sin(x))
    plt.plot(x, sin(2*x))
    plt.axhline(0, c='k', ls=':')
    plt.ylim([-1.5,1.5])
    ax = plt.gca()
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.set_xticks([])
    ax.set_xlabel('Time')
    ax.set_yticks([])
    ax.set_ylabel('Amplitude')
    plt.text(0, -1.2, 'g(t) = sin(2t)', c='C1')
    plt.savefig('/Users/oberrauch/Desktop/slides/pds2.pdf')

    plt.figure()
    plt.plot(x, 0.5*sin(x)+sin(2*x), c='C3')
    plt.axhline(0, c='k', ls=':')
    ax = plt.gca()
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.set_xticks([])
    ax.set_xlabel('Time')
    ax.set_yticks([])
    ax.set_ylabel('Amplitude')
    plt.text(0, -1.2, 'h(t) = f(t) + g(t)', c='C3')
    plt.savefig('/Users/oberrauch/Desktop/slides/pds3.pdf')
```
